# Task 3

```python
image_list, mask_list= create_data2('Lab3','MRI')
K = 3
#parameters
base_dense=8
batch_size=8
LR=0.00001
img_w,img_h,img_ch=240,240,2

#optimizer
optimizer =Adam(LR)

#metrics list
metric_list=[dice_coef,precision,recall,'binary_accuracy']
metric_str=['dice_coef','precision','recall','binary_accuracy']
#loss functions list
loss_func_list=[dice_coef_loss,'binary_crossentropy']

#global variable
lab_file='Lab3'
type_file='MRI'
```

```python
T = 2 # 2 cycles
for s in range(T):
    print('step number '+str(s))
    K = 3
    model_predictions=np.zeros((len(image_list), img_w, img_h,1))
    images_per_fold = int(len(image_list)/K)
    img_ch = 2
    output_path='../LAB5/'
    for fold in range(1):
        model=get_unet_batch(base_dense,img_w,img_h,2,dropout=True,dr=0.2)
        x_train,y_train,x_val,y_val=K_fold(image_list,mask_list,K_num=K,fold_num=fold,img_ch=1)
        if s == 0:
            autocontext_train=np.zeros_like(x_train)+0.5
            x_train=np.concatenate((x_train,autocontext_train),axis=-1)
            autocontext_val=np.zeros_like(x_val)+0.5
            x_val=np.concatenate((x_val,autocontext_val),axis=-1)
        else:
            y_pred=np.load(output_path+'step'+str(s-1)+'.npy')
            autocontext_train,autocontext_val=get_autocontext_fold2(y_pred,fold,K,images_per_fold)
            # Concatenate image data and posterior probabilities:
            x_train=np.concatenate((x_train,autocontext_train),axis=-1)
            x_val=np.concatenate((x_val,autocontext_val),axis=-1)
        model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=[metric_list[0]])
        model_hist=model.fit(x_train,y_train,batch_size=batch_size,epochs=100,validation_data=(x_val,y_val))           
        val_predictions=model.predict(x_val,batch_size=int(batch_size/2))
        model_predictions[(fold*images_per_fold):((fold+1)*images_per_fold)]=val_predictions
        np.save(output_path+'step'+str(s) +'.npy',model_predictions)
        
        plt.figure()
        plt.subplot(1,2,1)
        plt.title("Learning curve")
        plt.plot(model_hist.history["loss"], label="loss")
        plt.plot(model_hist.history["val_loss"], label="val_loss")
        plt.plot( np.argmin(model_hist.history["val_loss"]),
         np.min(model_hist.history["val_loss"]),
         marker="x", color="r", label="best model")
        plt.xlabel("Epochs")
        plt.ylabel("Loss Value")
        plt.legend();

        plt.subplot(1,2,2)
        plt.title("Learning curve")
        plt.plot(model_hist.history[metric_str[0]], label="accuracy")
        plt.plot(model_hist.history["val_"+metric_str[0]], label="val_accuracy")
        plt.plot( np.argmax(model_hist.history["val_"+metric_str[0]]),
         np.max(model_hist.history["val_"+metric_str[0]]),
         marker="x", color="r", label="best model")
        plt.xlabel("Epochs")
        plt.ylabel("Accuracy Value")
        plt.legend();
```
